# Plotting Functions:

Seaborn provides with many different kinds of functions to plot and vizualize different kinds of data. I'll talk about some of the most common ones and these can be divided into following categories:

## 1. **Visualizing statistical relationships**: 

One of the most important thing to do when you are playing around with data is to find out what sort of relationship does variables have with each other? Are they related to each other or not? etc.
Seaborn provies 3 functions to do so: 
    - replot(), 
    - lineplot() and 
    - scatterplot()

1.1  **replot()**

```python
sns.relplot(x="Weight",y="Height", hue="Sex", data= olympics);
```

1.2 **lineplot()**

```python
sns.lineplot(x="Age", y="Height", data=olympics)
```

```python
sns.lineplot(x="Age", y="Weight", data=olympics)
```

```python
sns.lineplot(x="Age", y="Height", hue="Season", data=olympics)
```

1.3 **scatterplot()**

```python
sns.scatterplot(x="Weight", y="Height", data=olympics);
```

## 2. Plotting with categorical data | categorical plot

   - Categorical scatterplots
     - stripplot() (with kind="strip"; the default)
     - swarmplot() (with kind="swarm")
   - Distributions of observations within categories
     - boxplot() (with kind="box")
     - violinplot() (with kind="violin")
     - boxenplot() (with kind="boxen")
   - Statistical estimation within categories
     - pointplot() (with kind="point")
     - barplot() (with kind="bar")
     - countplot() (with kind="count")


   2.1 Categorical scatterplots
    - stripplot() (with kind="strip"; the default)
    - swarmplot() (with kind="swarm")


```python
sns.stripplot(x="Medal", y="Age",hue="Season", data=olympics)
```

```python
sns.swarmplot(x="Medal", y="Age",hue="Season", data=olympics)
```

2.2 Categorical distribution plots:

    boxplot() (with kind="box")
    violinplot() (with kind="violin")
    boxenplot() (with kind="boxen")


```python
sns.boxplot(x="Medal", y="Age", data=olympics)
```

```python
sns.violinplot(x="Medal", y="Age", hue="Season", data=olympics)
```

```python
sns.boxenplot(x="Medal", y="Age", hue="Season", data=olympics)
```

2.3 Categorical estimate plots:

    pointplot() (with kind="point")
    barplot() (with kind="bar")
    countplot() (with kind="count")



```python
sns.barplot(x="Sex", y="Age", hue="Medal", data=olympics)
```

```python
sns.pointplot(x="Medal", y="Age", hue="Season", data=olympics)
```
