# Visualizing the distribution of a dataset

## Import libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

%matplotlib inline
```

```python
sns.set(color_codes=True)
```

When dealing with a dataset, often the first thing youâ€™ll want to do is to get a sense for how the variables are distributed. Seaborn gives you functions which can visualize how variables/ data is distributed, these can be divided broadly in 3 categories: 
- Plotting univariate distributions
- Plotting bivariate distributions
- Visualizing pairwise relationships in a dataset

## Plotting univariate distributions
The most convenient way to take a quick look at a univariate distribution in seaborn is the **distplot()** function.

```python
x = np.random.normal(size=100)
sns.distplot(x);
```

Letâ€™s remove the density curve and add a rug plot, which draws a small vertical tick at each observation. You can make the rug plot itself with the rugplot() function, but it is also available in distplot():

```python
sns.distplot(x, kde=False, rug=True);
```

You can also play around with number of bins to use while plotting your distribution:

```python
# with bins = 20
sns.distplot(x, bins=20, kde=False, rug=True);
```

```python
# with bins = 50
sns.distplot(x, bins=50, kde=False, rug=True);
```

## Kernel density estimation

You can use kdeplot() or even distplot().

```python
sns.distplot(x, hist=False, rug=True);
```

```python
sns.kdeplot(x, shade=True);
```

## Fitting parametric distributions

```python
x = np.random.gamma(6, size=200)
sns.distplot(x, kde=False, fit=stats.gamma);
```

## Plotting bivariate distributions
We can also visualize a bivariate distribution of two variables n seaborn. The easiest way to do this in seaborn is to just use the jointplot() function, which creates a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.

```python
# parameter values for 2 distributions
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
```

### Scatter plot

```python
sns.jointplot(x="x", y="y", data=df);
```

```python
# in matplotlib
plt.scatter(x="x", y="y", data=df)
```

### Hexbin plots

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```

### Kernel density estimation

```python
sns.jointplot(x="x", y="y", data=df, kind="kde");
```

## Visualizing pairwise relationships in a dataset
To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function.

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris);
```
